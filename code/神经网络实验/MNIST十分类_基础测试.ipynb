```python
import torch
import pandas as pd
import numpy as np
from scipy.io import loadmat
from sklearn.model_selection import train_test_split
from torch.utils.data import TensorDataset, DataLoader
from torch.optim import Optimizer
import torch.optim as optim
import torch.nn as nn
import torch.nn.functional as F
from useful_functions import *
from optimizer import *
from model import *
from MNIST_data_process import *
import matplotlib.pyplot as plt
from tqdm import tqdm
from training import *
from mlxtend.data import mnist_data
from accuracy_compute import *
```

```python
X, y = mnist_data()
X_tensor = torch.tensor(X, dtype=torch.float32)
y_tensor = torch.tensor(y, dtype=torch.long)
X_train, X_test, y_train, y_test = train_test_split(X_tensor, y_tensor, test_size=0.2, random_state=42)
model = MNISTClassifier_2layer_2()
criterion = nn.CrossEntropyLoss()
```

```python
optimizer = optim.SGD(model.parameters(), lr=1e-2)

# 训练模型
n_epochs = 500
for epoch in range(n_epochs):
    model.train()
    optimizer.zero_grad()  # 清空梯度
    output = model(X_train)  # 正向传播
    loss = criterion(output, y_train)  # 计算损失
    loss.backward()  # 反向传播
    optimizer.step()  # 更新参数

    accuracy=compute_accuracy_with_average_model([model],X_test,y_test)
    
    print(f'Epoch {epoch+1}/{n_epochs}, Loss: {loss.item()}, Test Accuracy: {accuracy * 100:.2f}%')

# 测试模型
model.eval()
with torch.no_grad():
    output = model(X_test)
    _, predicted = torch.max(output, 1)
    accuracy = (predicted == y_test).sum().item() / y_test.size(0)

print(f'Test Accuracy: {accuracy * 100:.2f}%')
```
